```python
import pandas as pd # importing or loading pandas library in this program for mathematical operation
```

```python
dataset=pd.read_csv("salary_data.csv") # attaching the datas which we collected from the customer
```

```python
dataset #checking the datas manualy
```

```python
independent = dataset[["YearsExperience"]] # assigning input data or split input data from dataset
```

```python
independent # verfing input data 
```

```python
dependant = dataset[["Salary"]] # split output data from dataset
```

```python
dependant # verfing output data
```

```python
from sklearn.model_selection import train_test_split # loading sklearn library for split train and test set
x_train,x_test,y_train,y_test=train_test_split(independent,dependant,test_size=0.30,random_state=0)# separting training set and data set from the input dataset
```

```python
from sklearn.linear_model import LinearRegression # loading linear regression function from sklearn library
regressor=LinearRegression()# assigning linear regression function to the regressor variable
regressor.fit(x_train,y_train)# assigning input and output parameter to the train set
```

```python
weight=regressor.coef_ # getting weight value from regressor
weight# displaying the weight value
```

```python
bias=regressor.intercept_# getting bias value from regressor
bias# displaying the bias value
```

```python
y_pred=regressor.predict(x_test) # testing the test set datas
```

```python
from sklearn.metrics import r2_score# loading r2 score from sklearn library
r_score=r2_score(y_test,y_pred)# testing r2 value with y test and y pred
```

```python
r_score # output of r score if it is >1 , it is good model
```

```python
import pickle # loading the pickle library for save the model
filename="finalized model_linear.sav" # file name with extaction of .sav for pickle library
```

```python
pickle.dump(regressor,open(filename,'wb')) #loading the file into pickle library
```

```python
loaded_model=pickle.load(open("finalized model_linear.sav",'rb'))#read the model
result=loaded_model.predict([[13]])
```
